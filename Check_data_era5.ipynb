# Get ERA5 data intersect with IOT

based on [omnicalculator](https://www.omnicalculator.com/physics/relative-humidity), relative humidity can be calculated using temperature and dew point temperature like this.

````python
import math

def calculate_relative_humidity(temp_celsius, dew_point_celsius):
    # Constants for the formula
    A = 17.625
    B = 243.04

    # Calculate saturation vapor pressure for air temperature
    sat_vapor_pressure_temp = math.exp((A * temp_celsius) / (B + temp_celsius))

    # Calculate saturation vapor pressure for dew point temperature
    sat_vapor_pressure_dew_point = math.exp((A * dew_point_celsius) / (B + dew_point_celsius))

    # Calculate relative humidity
    relative_humidity = 100 * (sat_vapor_pressure_dew_point / sat_vapor_pressure_temp)

    return relative_humidity
````

````python
for data in list_sta:
  print (data[45:51]+'_iot')
  # read data
  globals()[data[45:51]+'_iot']=pd.read_csv(data)

  # change column type into datetime
  globals()[data[45:51]+'_iot']['time']=pd.to_datetime(globals()[data[45:51]+'_iot']['time'], errors='coerce')

  # Define the complete date range (assuming hourly frequency for this example)
  globals()[data[45:51]+'_iot'].set_index('time', inplace=True)
  start_date = globals()[data[45:51]+'_iot'].index.min()
  end_date = globals()[data[45:51]+'_iot'].index.max()
  complete_date_range = pd.date_range(start=start_date, end=end_date, freq='H')

  # Reindex the DataFrame to the complete date range, filling missing values with NaN
  globals()[data[45:51]+'_iot'] = globals()[data[45:51]+'_iot'].reindex(complete_date_range).reset_index().rename(columns={'index': 'time'})

  # Separate the columns to apply different fill methods
  globals()[data[45:51]+'_iot']['rainfall'] = globals()[data[45:51]+'_iot']['rainfall'].fillna(0)
  globals()[data[45:51]+'_iot']['altitude'] = globals()[data[45:51]+'_iot']['altitude'].fillna(method='ffill')
  globals()[data[45:51]+'_iot']['pressure'] = globals()[data[45:51]+'_iot']['pressure'].fillna(method='ffill')*100
  globals()[data[45:51]+'_iot']['temperature'] = globals()[data[45:51]+'_iot']['temperature'].fillna(method='ffill')
  globals()[data[45:51]+'_iot']['humidity'] = globals()[data[45:51]+'_iot']['humidity'].fillna(method='ffill')
  try:
    globals()[data[45:51]+'_iot']['intensity'] = globals()[data[45:51]+'_iot']['intensity'].fillna(method='ffill')
  except:
    print(data[45:51]+'_iot has no intensity sensor')

  # filling a null values using fillna()
  globals()[data[45:51]+'_iot']['device'].fillna(globals()[data[45:51]+'_iot']['device'].unique()[0], inplace=True)
  globals()[data[45:51]+'_iot']['device_id'].fillna(globals()[data[45:51]+'_iot']['device_id'].unique()[0], inplace=True)
  globals()[data[45:51]+'_iot']['farm_area_code'].fillna(globals()[data[45:51]+'_iot']['farm_area_code'].unique()[0], inplace=True)
  try:
    globals()[data[45:51]+'_iot']['_measurement'].fillna(globals()[data[45:51]+'_iot']['_measurement'].unique()[0], inplace=True)
    globals()[data[45:51]+'_iot']['farm_id'].fillna(globals()[data[45:51]+'_iot']['farm_id'].unique()[0], inplace=True)
  except:
    print(data[45:51]+'_iot has no attribute _measurement and farm_id')

  # add column for 6 hourly time
  globals()[data[45:51]+'_iot']['6h']=[
    7 if (1 < x.hour <= 7) else
    13 if (7 < x.hour <= 13) else
    19 if (13 < x.hour <= 19) else
    1 if ((19 < x.hour <= 23) or (0 <= x.hour <= 1)) else None  # Ensure all conditions are handled
    for x in globals()[data[45:51]+'_iot']['time']
  ]
  globals()[data[45:51]+'_iot']['date']=globals()[data[45:51]+'_iot']['time'].dt.date
  globals()[data[45:51]+'_iot']['time_6h']=[str(a)+' '+"{:02}".format(int(b))+':00' for a, b in zip(globals()[data[45:51]+'_iot']['date'], globals()[data[45:51]+'_iot']['6h'])]
  globals()[data[45:51]+'_iot']['time_6h']=pd.to_datetime(globals()[data[45:51]+'_iot']['time_6h'], errors='coerce')
  globals()[data[45:51]+'_iot'].drop(['date', '6h'], axis=1, inplace=True)
````

## Change time interval into 6 hourly

Data era5 has 6 time interval for 00, 06, 12, and 18 UTC or 07, 13, 19, and 01 WIB (UTC+7). Then we will change the IOT data with 6 hours interval.

````python
 for data in list_sta:
  print (data[45:51]+'_iot2')
  # set data 6 hourly
  try:
    data2=globals()[data[45:51]+'_iot'].groupby(['time_6h','device','device_id']).agg({'rainfall':'sum', 'altitude':'mean','pressure':'mean',
                                    'temperature':'mean','humidity':'mean', 'intensity':'mean'})
  except:
    data2=globals()[data[45:51]+'_iot'].groupby(['time_6h','device','device_id']).agg({'rainfall':'sum', 'altitude':'mean','pressure':'mean',
                                    'temperature':'mean','humidity':'mean'})

  # Aggregation for mode fields using apply
  mode_fields = ['farm_area_code']

  data3 = globals()[data[45:51]+'_iot'].groupby(['time_6h','device','device_id'])[mode_fields].apply(lambda x: x.mode().iloc[0])

  # Merging both aggregated parts
  final_aggregated_data = pd.concat([data3, data2], axis=1)
  globals()[data[45:51]+'_iot2']=final_aggregated_data.reset_index(['device','device_id']).sort_index(ascending=True).reset_index()
  globals()[data[45:51]+'_iot2'].rename({'time_6h': 'time'}, axis=1, inplace=True)
  globals()[data[45:51]+'_iot2'].set_index('time', inplace=True)
````

````python
# Vectorized function to calculate relative humidity
def calculate_relative_humidity(temp_celsius, dew_point_celsius):
    # Constants for the formula
    A = 17.625
    B = 243.04

    # Calculate saturation vapor pressure for air temperature
    sat_vapor_pressure_temp = np.exp((A * temp_celsius) / (B + temp_celsius))

    # Calculate saturation vapor pressure for dew point temperature
    sat_vapor_pressure_dew_point = np.exp((A * dew_point_celsius) / (B + dew_point_celsius))

    # Calculate relative humidity
    relative_humidity = 100 * (sat_vapor_pressure_dew_point / sat_vapor_pressure_temp)

    return relative_humidity
````
